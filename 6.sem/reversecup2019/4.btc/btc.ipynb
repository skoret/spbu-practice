```python
import os, binascii, base58
```

### decode wif private keys

```python
def wif_to_pkh(wif):
    ver_fullkey_sum = binascii.hexlify(base58.b58decode(wif)).decode()
    return ver_fullkey_sum[2:-8]

def wif_to_pk(wif):
    return binascii.unhexlify(wif_to_pkh(wif))
```

```python
wif = '5J6CkuTkeQ611csdhSYdG3ds3fQyiKDPFzcfuVmjZXTV8Zh8p6o'
form = '{:<18} {}'
print(form.format('wif:', wif))
print(form.format('private key hex:', wif_to_pkh(wif)))
print(form.format('private key ascii:', wif_to_pk(wif).decode()))
```

### parse file

```python
file = os.path.join(os.getcwd(), 'noname')
ns = os.path.join(os.getcwd(), 'n1.s')
```

```python
def wif_parse(file, ind):
    with open(file, 'r') as wifs:
        for wif in wifs:
            if ind <= 0:
                break
            ind -= 1
            pk = wif_to_pk(wif).decode()
            link = pk[:16].replace('#', ' ').replace('$', ' ')
            code = pk[16:].replace('#', ' ').strip()
            line = '{} | {} | {}'.format(pk, link, code)
            print(line)
```

```python
def wif_to_asm(inf, outf, ind):
    with open(inf, 'r') as i, open(outf, 'w') as o:
        for wif in i:
            if ind <= 0:
                break
            ind -= 1
            pk = wif_to_pk(wif).decode()[16:].replace('#', ' ').strip()
            o.write(pk + '\n')
            print(pk)
```

```python
wif_parse(file, 205)
```

```python
wif_to_asm(file, ns, 194)
```
